```python
#Importações das bibliotecas para manipulação, tratamento e visualização dos dados
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

#Importações de funções e variáveis criadas
from src.config import DADOS_LIMPOS
from src.graficos import SCATTER_ALPHA, PALETTE

sns.set_theme(palette=PALETTE, style='white')
```

```python
df = pd.read_parquet(DADOS_LIMPOS)

df
```

```python
df.info()
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(figsize=(12,8))

sns.scatterplot(data=df, x='longitude', y='latitude', hue='ocean_proximity',ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))

sns.scatterplot(data=df, x='longitude', y='latitude', hue='median_income_cat', palette=PALETTE, ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))

norm_median_house_value = plt.Normalize(df['median_house_value'].min(), df['median_house_value'].max())
sm_median_house_value = plt.cm.ScalarMappable(norm= norm_median_house_value, cmap=PALETTE)

sns.scatterplot(data=df, x='longitude', y='latitude', hue='median_house_value', palette=PALETTE, ax=ax)

ax.get_legend().remove()

fig.colorbar(sm_median_house_value, ax=ax)

plt.show()
```
